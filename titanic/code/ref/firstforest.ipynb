```python
import pandas as pd
import numpy as np
import csv as csv
from sklearn.ensemble import RandomForestClassifier
```

```python
# Data cleanup
# TRAIN DATA
train_df = pd.read_csv('../input/train.csv', header=0)    
```

```python
train_df.info()
```

```python
train_df['Gender'] = train_df['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
```

```python
if len(train_df.Embarked[ train_df.Embarked.isnull() ]) > 0:
    train_df.Embarked[ train_df.Embarked.isnull() ] = train_df.Embarked.dropna().mode().values

Ports = list(enumerate(np.unique(train_df['Embarked']))) 
```

```python
Ports_dict = { name : i for i, name in Ports }   
```

```python
train_df.Embarked = train_df.Embarked.map( lambda x: Ports_dict[x]).astype(int)
```

```python
median_age = train_df['Age'].dropna().median()
if len(train_df.Age[ train_df.Age.isnull() ]) > 0:
    train_df.loc[ (train_df.Age.isnull()), 'Age'] = median_age
```

```python
train_x = train_df.iloc[:800, 1:]
train_y = train_df.iloc[:800, 0]

test_x = train_df.iloc[800:, 1:]
test_y = train_df.iloc[800:, 0]
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LogisticRegressionCV
from sklearn.ensemble.bagging import DecisionTreeClassifier

from sklearn.cross_validation import cross_val_score, train_test_split

clf = [SVC, AdaBoostClassifier, RandomForestClassifier, LogisticRegression, LogisticRegressionCV, DecisionTreeClassifier]

models = [c().fit(train_df.drop('Survived', axis=1), train_df.Survived) for c in clf]
res = np.array([model.predict(test_df) for model in models])
```

```python
predict = (res.mean(axis=0) > 0.5).astype(int)
predict.shape
```

```python
predictions_file = open("myfirstforest.csv", "wb")
open_file_object = csv.writer(predictions_file)
open_file_object.writerow(["PassengerId","Survived"])
open_file_object.writerows(zip(ids, predict))
predictions_file.close()
print 'Done.'
```
